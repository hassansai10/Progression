## 25=0 for lung_cancer when we 25/276 we get 0.09 so 9% of the whole dataset is filled with data of patients that do not have lung cancer so we need over sampling to match both of the densities, or you can say to make both densities of True and Flase Equal or Somewhat Equal

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
```

```python
X_before_Over_Sampling = data.drop(columns=['LUNG_CANCER'])
Y_before_Over_Sampling = data['LUNG_CANCER']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X_before_Over_Sampling,Y_before_Over_Sampling,random_state=42)
```

```python
minority_sampling = SMOTE(random_state=42)
```

```python
X,Y = minority_sampling.fit_resample(X_train,Y_train)
```

```python
print (f'Original Destribution {Counter(Y_train)}')
print (f'Resample Destribution {Counter(Y)}')
```

```python
preprocess = ColumnTransformer(transformers=[('num',StandardScaler(),['GENDER', 'AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY','PEER_PRESSURE', 'CHRONIC DISEASE', 'FATIGUE ', 'ALLERGY ', 'WHEEZING','ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH','SWALLOWING DIFFICULTY', 'CHEST PAIN'])])
```

```python
Model = pipeline(steps=[('preprocessor',preprocess),('polynomial_features',PolynomialFeatures(degree=1)),('regressor',LogisticRegression(solver='saga',max_iter=1000,random_state=42))])
```

```python
GridSearch_params = {'regressor__C':[0.1,1,10,100],'regressor__max_iter':[100,1000,2000],'polynomial_features__degree':[1,2,3],'regressor__solver':['saga','liblinear']}
```

```python
GridSearch = GridSearchCV(Model,GridSearch_params,cv=5,scoring='accuracy',n_jobs=-1,verbose=2)
GridSearch.fit(X,Y)
```

```python
print (f'Best Parameter = {GridSearch.best_params_}')
```

```python
best_model = GridSearch.best_estimator_
```

```python
cross_validation = cross_val_score(best_model,X,Y,cv=5,scoring='accuracy',n_jobs=-1,verbose=2)
```

```python
print (f'Model Accuracy {cross_validation.mean()}')
```

```python
predict = best_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
Accuracy = accuracy_score(predict,Y_test)
```

```python
print (f'Accuracy = {Accuracy*100:.2f}')
```
